### Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### 
This notebook compares the performance of Logistic Regression,
Random Forest and XGBoost models developed during the project.

The objective is to identify the most suitable model for customer churn
prediction while considering both predictive performance and
interpretability.

### Create Model Comparison Table

```python
comparison_df = pd.DataFrame({
    "Model": [
        "Logistic Regression",
        "Random Forest",
        "XGBoost"
    ],

    "Accuracy": [
        0.89,
        0.9995,
        0.9999
    ],

    "Precision": [
        0.92,
        0.9999,
        1.00
    ],

    "Recall": [
        0.89,
        0.9992,
        0.9999
    ],

    "F1 Score": [
        0.904,
        0.9996,
        0.9999
    ],

    "ROC-AUC": [
        0.96,
        0.999999,
        0.999999
    ]
})

comparison_df
```

### 
The comparison table summarises the predictive performance of all three
models across multiple evaluation metrics.

Using multiple metrics provides a more comprehensive assessment of model
performance than accuracy alone.

### Accuracy Comparison

```python
comparison_df[["Model","Accuracy"]]
```

### 
XGBoost achieved the highest accuracy, closely followed by Random Forest.

Both tree-based models substantially outperformed Logistic Regression,
suggesting that customer churn behaviour contains important non-linear
relationships.

### Precision Comparison

```python
comparison_df[["Model","Precision"]]
```

### 
XGBoost achieved perfect precision, while Random Forest also achieved
near-perfect precision.

This indicates that both models generated very few false positive churn
predictions.

### Recall Comparison

```python
comparison_df[["Model","Recall"]]
```

### 
XGBoost achieved the highest recall, successfully identifying almost all
customers who eventually churned.

High recall is particularly valuable in customer retention problems
because missed churners represent lost revenue opportunities.

### F1 Score Comparison

```python
comparison_df[["Model","F1 Score"]]
```

### 
XGBoost achieved the highest F1 score, indicating the strongest balance
between precision and recall.

This suggests that XGBoost provided the most effective overall churn
classification performance.

### ROC-AUC Comparison

```python
comparison_df[["Model","ROC-AUC"]]
```

### 
Both Random Forest and XGBoost achieved near-perfect ROC-AUC scores,
demonstrating exceptional ability to distinguish between churned and
retained customers across all classification thresholds.

### Performance Visualization

```python
comparison_df.plot(
    x="Model",
    y=["Accuracy","Precision","Recall","F1 Score","ROC-AUC"],
    kind="bar",
    figsize=(10,6)
)

plt.title("Model Performance Comparison")
plt.ylabel("Score")
plt.show()
```

### 
The performance visualization demonstrates a clear improvement in model performance as model complexity increased.

Logistic Regression provided a strong and interpretable baseline model, achieving approximately 89% accuracy and a ROC-AUC of 0.96.

Both Random Forest and XGBoost substantially outperformed Logistic Regression across all evaluation metrics, achieving near-perfect classification performance.

XGBoost emerged as the strongest overall model, achieving the highest accuracy, precision, recall, F1 score and ROC AUC, although the performance difference relative to Random Forest was very small.

These results suggest that customer churn in this dataset is influenced by complex non-linear relationships that tree-based ensemble methods are better able to capture than linear models.

```python
print(
    "Interpretability vs Performance Analysis"
)
```

###

This highlights a common trade-off in machine learning between model
