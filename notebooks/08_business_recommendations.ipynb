The objective of this notebook is to convert analytical findings into practical business strategies aimed at reducing customer churn and improving customer retention.

### Key Drivers of Churn

```python
key_drivers = [
    "Contract Length_Monthly",
    "Support Calls",
    "Total Spend",
    "Payment Delay",
    "Last Interaction"
]

print(key_drivers)
```

### 
Feature importance analysis across Logistic Regression, Random Forest
and XGBoost consistently identified contract type, customer support
activity, spending behaviour, payment delays and customer engagement as
the primary drivers of churn.

### Recommendation 1: Target Monthly Customers

```python
print(
    "Focus retention efforts on customers with monthly contracts."
)
```

### 
Customers on monthly contracts exhibited substantially higher churn risk
than customers on annual or quarterly contracts.

The business should consider offering incentives such as discounts,
reward programmes or contract upgrades to encourage longer-term
commitment.


### Recommendation 2: Monitor Support Calls

```python
print(
    "Monitor customers with increasing support call volumes."
)
```

### 
Support Calls emerged as one of the strongest predictors of customer
churn across all models.

A large number of support interactions may indicate customer
dissatisfaction or unresolved issues.

Customers with repeated support requests should be prioritised for
proactive intervention.

### Recommendation 3: Monitor Payment Delays

```python
print(
    "Flag customers with increasing payment delays."
)
```

### 
Payment Delay was consistently identified as a major churn predictor.

Customers exhibiting increasing payment delays may be experiencing financial difficulties or reduced engagement with the service.

Early intervention may prevent customer loss.

### Recommendation 4: Protect High-Value Customers

```python
print(
    "Prioritise high-value customers for retention programmes."
)
```

### 
Customers with higher spending levels contributed significantly to business revenue.

Retention resources should therefore be prioritised toward high-value customers exhibiting elevated churn risk.

### Recommendation 5: Improve Customer Engagement

```python
print(
    "Increase engagement with inactive customers."
)
```

### 
Longer periods since the last customer interaction were associated with
higher churn rates.

Regular engagement campaigns, loyalty programmes and personalised
communication may reduce customer attrition.

### Early Warning System

```python
print(
    "Use the model to continuously identify customers at high risk of churn.")
```

### 
The XGBoost model could be deployed as an early warning system that
scores customers according to their probability of churn.

Customers identified as high risk could automatically be referred to
customer retention teams for intervention.


### Potential Business Impact

```python
print(
    "Estimate potential business value."
)
```

### 
Reducing customer churn can significantly increase customer lifetime
value and reduce acquisition costs.

Because acquiring new customers is generally more expensive than
retaining existing customers, effective churn prediction can generate
substantial financial benefits.

### Limitations

```python
print(
    "Document project limitations."
)
```

### 
The near-perfect predictive performance achieved by Random Forest and
XGBoost is uncommon in real-world churn datasets.

Additional validation using external or future customer data would be
necessary before deploying the models in production environments.

Furthermore, model performance should be monitored continuously to
ensure long-term generalisation.

### Final Business Recommendation

```python
print(
    "Deploy XGBoost with monitoring and retention workflows."
)
```

###
XGBoost demonstrated the strongest predictive performance and would be
the preferred deployment model for churn prediction.

However, deployment should be accompanied by ongoing monitoring,
periodic retraining and validation to ensure reliable long-term
performance.
